I decided to recreate the visualization titled "Per capita CO2 emissions" from a website called Our World in Data. This dataset measures the average annual emissions per person for a country or region. These values are calculated by dividing the total annual emissions of the country or region by its population.

There were three main sources of data used by the website in order to compile this information. The first was the HYDE database which gives history databases of different environments, the second source was Gapminder and the third source was the UN World Population Prospect.

I chose this visualization because I am very interested in topics regarding carbon footprints. I thought it would be interesting to see how the visualization is able to calculate a carbon emission value for the average person in each country or region. I was also interested to see how much higher the per capita co2 emission for the United States was compared to other countries or regions.

Another intersting factor of this visualization was the timespan. The data ranges from 1750-2022 and is recalculated every year. There is also an interactive button that allows you to view which countries were leading in the Per capita CO2 emissions.

Here is the link to the original visualization: https://ourworldindata.org/co2-and-greenhouse-gas-emissions



# Visualization recreation in Python using Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the csv file
co2_emissions = pd.read_csv('co-emissions-per-capita.csv')
```

```python
#filtered out data for the specific countries used in Our World in Data
entities = ['United States', 'United Kingdom', 'China', 'India', 'World']
filtered_co2 = co2_emissions[co2_emissions['Entity'].isin(entities)]
```

```python
#indexes the rows by year, columns represent each entity and cells contain each annual co2 emission
pivoted_co2 = filtered_co2.pivot(index='Year', columns='Entity', values='Annual CO₂ emissions (per capita)')
```

```python
#was running into errors here so ended up using chatgpt
#used plt.plot
plt.figure(figsize=(15, 7))
for entity in entities:
    plt.plot(pivoted_co2.index, pivoted_co2[entity], label=entity)

plt.title('Per Capita CO₂ Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('Annual CO₂ emissions (per capita)')
plt.legend()
plt.grid(True)
plt.show()
```
